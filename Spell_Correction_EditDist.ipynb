```python
import nltk
# nltk.download()
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import string  
```

```python
pun = string.punctuation 
stopwords = stopwords.words("english")
stopwords += list(pun)  # Creating a list of words and punctuations that have less relevance
print(stopwords)
lemmatizer = WordNetLemmatizer() 
```

```python
f = open("Sample Texts/alice_correct.txt",'r')
text = f.read()
f.close()

f = open("Sample Texts/alice_misspelled.txt",'r')
wrongtext = f.read()
f.close()
```

```python
def vocablist(text,stopwords) :
  word_tokens = word_tokenize(text)
  tokens = []
  for word in word_tokens :
    if word.lower() not in stopwords :
      tokens.append(word) 
  for word in tokens :
    lemmatizer.lemmatize(word)
  types = list(set(tokens))
  return types

print("Vocabilary List :")
print(vocablist(text,stopwords))
print()
```

```python
def mispelledlist(text,wrongtext,stopwords) :
  correctlist = vocablist(text,stopwords)
  wronglist = vocablist(wrongtext,stopwords)
  mispelled = []
  for i in wronglist :
    if i not in correctlist :
      mispelled.append(i)
  return mispelled

print("Mis-spelled list :")
print(mispelledlist(text,wrongtext,stopwords))
```

```python
def editDistance(str1, str2, m, n): 
  ed = [[0 for x in range(n + 1)] for x in range(m + 1)] 
  for i in range(m + 1): 
    for j in range(n + 1): 
      if i == 0: 
        ed[i][j] = j 
      elif j == 0: 
        ed[i][j] = i  
      elif str1[i-1] == str2[j-1]: 
        ed[i][j] = ed[i-1][j-1] 
      else: 
        ed[i][j] = 1 + min(ed[i][j-1], ed[i-1][j] , ed[i-1][j-1])
  return ed[m][n] 

def spellCheck(l1,l2) : # l1 = wl  l2 = cl
  spell = []
  for i in range(len(l1)) : 
    nearlycorrect =["Wrong word :"+l1[i]]
    for j in range(len(l2)) :     
      # print(i)
      ed = editDistance(l1[i],l2[j],len(l1[i]),len(l2[j]))
      # print(ed)
      if  ed<= 2 :
        nearlycorrect.append(l2[j])
    spell.append(nearlycorrect)
  return spell

print(spellCheck(mispelledlist(text,wrongtext,stopwords),vocablist(text,stopwords)))
```
