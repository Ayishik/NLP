```python
def deletion_matrix(count_edit) :                 #Confusion matrix for delete
  character = []
  for edit in count_edit :
    if edit[0][0] not in character :
      character.append(edit[0][0])
    if edit[0][1] not in character :
      character.append(edit[0][1])
  del_mat = [[1]*(len(character)+1) for i in range(len(character)+1)]
  for i in range(1,len(character)+1) :
    del_mat[i][0] = character[i-1]
    del_mat[0][i] = character[i-1]
  for edit in count_edit :
    if len(edit[0][0]) < len(edit[0][1]) :
      # print(edit[1])
      del_mat[character.index(edit[0][1])+1][character.index(edit[0][0])+1] = edit[1]
      # print(del_mat[character.index(edit[0][1])+1][character.index(edit[0][0])+1])
  # print(np.matrix(del_mat))
  summation = [0]
  for i in range(1,len(character)+1) :
    add = 0
    for j in range(1,len(character)+1) :
      del_mat[i][j] = int(del_mat[i][j]) 
      # print(del_mat[i][j])
      add += del_mat[i][j]
    summation.append(add)
  # print(summation)
  for i in range(1,len(character)+1) :
    for j in range(1,len(character)+1) :
      if summation[i] != 0 :
        del_mat[i][j]/= (summation[i]+len(character))
  # print(np.matrix(del_mat))
  return del_mat

def insertion_matrix(count_edit) :                          #Confusion matrix for insertion
  character = []
  for edit in count_edit :
    if edit[0][0] not in character :
      character.append(edit[0][0])
    if edit[0][1] not in character :
      character.append(edit[0][1])
  ins_mat = [[1]*(len(character)+1) for i in range(len(character)+1)]
  for i in range(1,len(character)+1) :
    ins_mat[i][0] = character[i-1]
    ins_mat[0][i] = character[i-1]
  for edit in count_edit :
    if len(edit[0][0]) > len(edit[0][1]) :
      # print(edit[1])
      ins_mat[character.index(edit[0][1])+1][character.index(edit[0][0])+1] = edit[1]
      # print(del_mat[character.index(edit[0][1])+1][character.index(edit[0][0])+1])
  # print(np.matrix(del_mat))
  summation = [0]
  for i in range(1,len(character)+1) :
    add = 0
    for j in range(1,len(character)+1) :
      ins_mat[i][j] = int(ins_mat[i][j]) 
      # print(del_mat[i][j])
      add += ins_mat[i][j]
    summation.append(add)
  # print(summation)
  for i in range(1,len(character)+1) :
    for j in range(1,len(character)+1) :
      if summation[i] != 0 :
        ins_mat[i][j]/= (summation[i]+len(character))
        # print(ins_mat[i][j])
  # print(np.matrix(ins_mat))
  # print(ins_mat[1])
  return ins_mat

def substitution_matrix(count_edit) :                          #Confusion matrix for substitutuion
  character = []
  for edit in count_edit :
    if edit[0][0] not in character :
      character.append(edit[0][0])
    if edit[0][1] not in character :
      character.append(edit[0][1])
  sub_mat = [[1]*(len(character)+1) for i in range(len(character)+1)]
  for i in range(1,len(character)+1) :
    sub_mat[i][0] = character[i-1]
    sub_mat[0][i] = character[i-1]
  for edit in count_edit :
    if len(edit[0][0]) == len(edit[0][1]) :
      # print(edit[1])
      sub_mat[character.index(edit[0][1])+1][character.index(edit[0][0])+1] = edit[1]
      # print(del_mat[character.index(edit[0][1])+1][character.index(edit[0][0])+1])
  # print(np.matrix(del_mat))
  summation = [0]
  for i in range(1,len(character)+1) :
    add = 0
    for j in range(1,len(character)+1) :
      sub_mat[i][j] = int(sub_mat[i][j]) 
      # print(del_mat[i][j])
      add += sub_mat[i][j]
    summation.append(add)
  # print(summation)
  for i in range(1,len(character)+1) :
    for j in range(1,len(character)+1) :
      if summation[i] != 0 :
        sub_mat[i][j]/= (summation[i]+len(character))
        # print(ins_mat[i][j])
  # print(np.matrix(ins_mat))
  # print(ins_mat[1])
  return sub_mat

del_mat = deletion_matrix(count_edit)
ins_mat = insertion_matrix(count_edit)
sub_mat = substitution_matrix(count_edit)
print(np.matrix(del_mat))
print('\n')
print(np.matrix(ins_mat))
print('\n')
print(np.matrix(sub_mat))
```

```python
def shorten(text) :
  word_token = word_tokenize(text)
  k = word_token.index("II")
  token = []
  for i in range(0,k-1) :
    token.append(word_token[i])
  # print(token)
  return ' '.join(token)

text_ch1 = shorten(text)
wrongtext_ch1 = shorten(wrongtext)
```

```python
def vocablist(text,stopwords) :                    # Creating a list of words in vocabulary for the chapter
  word_tokens = word_tokenize(text)
  tokens = []
  for word in word_tokens :
    if word.lower() not in stopwords :
      tokens.append(word) 
  types = list(set(tokens))
  return types


def mispelledlist(text,wrongtext,stopwords) :     #Finds the wrong words by comparing correct words
  correctlist = vocablist(text,stopwords)
  wronglist = vocablist(wrongtext,stopwords)
  mispelled = []
  for i in wronglist :
    if i not in correctlist :
      mispelled.append(i)
  return mispelled

def editDistance(str1, str2, m, n): 
  ed = [[0 for x in range(n + 1)] for x in range(m + 1)] 
  for i in range(m + 1): 
    for j in range(n + 1): 
      if i == 0: 
        ed[i][j] = j 
      elif j == 0: 
        ed[i][j] = i  
      elif str1[i-1] == str2[j-1]: 
        ed[i][j] = ed[i-1][j-1] 
      else: 
        ed[i][j] = 1 + min(ed[i][j-1], ed[i-1][j] , ed[i-1][j-1])
  return ed[m][n] 

def spellCheck(l1,l2) : # l1 = wl  l2 = cl      #Using edit distance to decrease the number of errors to check using probability
  spell = []
  for i in range(len(l1)) :
    nearlycorrect =["Wrong word :"+l1[i]]
    for j in range(len(l2)) :     
      # print(i)
      ed = editDistance(l1[i],l2[j],len(l1[i]),len(l2[j]))
      # print(ed)
      if  ed<= 1 :
        nearlycorrect.append(l2[j])
    spell.append(nearlycorrect)
  return spell

print(vocablist(text_ch1,stopwords))
```

```python
def counts(word,text) :
  count = 0
  # word_token = word_tokenize(text)
  for words in text :
    if  words == word :
      count += 1
  return count

def maxArr(a) :
  max = a[0]
  for i in a :
    if i > max :
      max = i 
  for i in range(len(a)) :
    if a[i] == max :
      break
  return i
```

```python
def errorIs(word1,word2) :          #Tells whether the error is insertion, deletion or substitutuion 
  # print(len(word1),len(word2))
  if len(word1) == len(word2) :
    return 0
  if len(word1) > len(word2) :
    return 1
  if len(word1) < len(word2) :
    return 2

def sub(word1,word2) :         #Fixes substitution mistakes
  l1 = list(word1)
  l2 = list(word2)
  subs = []
  for i in range(len(word1)) :
    if word1[i] != word2[i] :
      subs.append(word1[i])
      subs.append(word2[i])
      break
  return subs

def ins(word1,word2) :        #Fixes insertion mistakes
  inst = []
  if len(word1) == 1 :
    inst.append(word1[0]+' ')
    inst.append(word2[-1])
  if len(word2) == 1 :
    inst.append(word1[-1])
    inst.append(word2[0]+' ')
  for i in range(1,len(word2)) :
    if word1[i-1] != word2[i-1] :
      inst.append(word1[i-1])
      inst.append(word2[i-1]+word2[i])
      break
  if len(inst) == 0 :
    inst.append(word1[-1])
    inst.append(word2[-2]+word2[-1])
  return inst

def dele(word1,word2) :     #Fixes deletion mistakes
  # print(word1,word2)
  inst = []
  if len(word1) == 1 :
    inst.append(word1[0]+' ')
    inst.append(word2[-1])
  if len(word2) == 1 :
    inst.append(word1[-1])
    inst.append(word2[0]+' ')
  for i in range(1,len(word1)) :
    if word2[i-1] != word1[i-1] :      
      inst.append(word1[i-1]+word1[i])
      inst.append(word2[i-1])
      break
  if len(inst) == 0 :
    inst.append(word1[-2]+word1[-1])
    inst.append(word2[-1])
  return inst
```
