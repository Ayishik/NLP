```
import numpy as np
from tqdm import tqdm

vectors = {}
word2vec = {}
words = []
with open('glove.6B.300d.txt') as f:
    for line in tqdm(f, total=400000):
        word, vector = line.split(maxsplit=1)
        words.append(word)
        v = np.fromstring(vector, sep=' ', dtype='float32')
        vectors[word] = v 
        word2vec[word] = v / np.linalg.norm(v)
```

```
def merge(arr, l, m, r):                  # This merge sort is used to compare the tuples of all potential words              
    n1 = m - l + 1
    n2 = r- m 
  
    L = [0] * (n1) 
    R = [0] * (n2) 
  
    for i in range(0 , n1): 
        L[i] = arr[l + i] 
  
    for j in range(0 , n2): 
        R[j] = arr[m + 1 + j] 
  
    
    i = 0     
    j = 0     
    k = l     
  
    while i < n1 and j < n2 : 
        if L[i][-1] >= R[j][-1]: 
            arr[k] = L[i] 
            i += 1
        else: 
            arr[k] = R[j] 
            j += 1
        k += 1
  
   
    while i < n1: 
        arr[k] = L[i] 
        i += 1
        k += 1
  
    while j < n2: 
        arr[k] = R[j] 
        j += 1
        k += 1
  
def mergeSort(arr,l,r): 
    if l < r: 
  
        m = (l+(r-1))//2

        mergeSort(arr, l, m) 
        mergeSort(arr, m+1, r) 
        merge(arr, l, m, r)
```

```
def cosineSimilarity(df,w1,w2) :                      # Finds the cosine similarity, that is, higher the similarity(near to 1) more similar the words are
  dotProduct = (df[w1] * df[w2]).sum()
  modW1 = ((df[w1] ** 2).sum()) ** 0.5
  modW2 = ((df[w2] ** 2).sum()) ** 0.5
  
  return dotProduct/modW1/modW2
```

```
print(cosineSimilarity(df,'japan','tokyo'))           # Two exmples shows how analogies can be found
print(cosineSimilarity(df,'india','delhi'))           # The cosine similarity of japan->tokyo is very near to india->delhi
```

```
def findAnalogy(df,W1,w1,w2,word = words) :

  assumedVec = cosineSimilarity(df,w1,w2)            # Finding cosine similarity of the known pair

  potential = []
  for word in words :                                # Comparing the cosine similarities of all the words with the one part of the unknown pair in Glove to find the nearest pair
    if word != W1 :
      cosineSim = cosineSimilarity(df,W1,word)
      t = (word,cosineSim)                           # Storing the word and tuple to use later
      potential.append(t)

  mergeSort(potential,0,len(potential)-1)            # Sorting to get the most similar words

  for i in range(10) :
    print(potential[i][0])
```

```
findAnalogy(df,'king','man','woman')
```

```
findAnalogy(df,'russia','china','beijing')
```

```
findAnalogy(df,'uncle','brother','sister')
```

```
def findNearest(df,w1,w2,words = words) :
  vecSum = (df[w1] + df[w2])                   # Vector addition in 300D of the two words
  modVecSum = (vecSum ** 2).sum()              # Calculating the mod of 300D vector for cosine similarity which is done later

  potential = []
  for word in words :
    if word not in [w1,w2] :                   # Avoiding the two words as these will give a ver good similarity
      cosSim = (df[word] * vecSum).sum()/modVecSum/(((df[word] ** 2).sum()) ** 0.5)      # cosSim = dot product of word w and vecSum / mod(vecSum) / mod(word w vector)
      t = (word,cosSim)
      potential.append(t)

  mergeSort(potential,0,len(potential)-1)

  for i in range(10) :
    print(potential[i][0])
```

```
findNearest(df,'russian','river')
```

```
findNearest(df,'india','holy')
```

```
findNearest(df,'german','airlines')
```
